PCA, ou principal component analysis  uma tcnica para reduzir a dimenso de um conjunto de dados preservando, de certa forma, suas propriedades.

Executar essa tcnica na mo ou sem ajuda de funes prontas  um bom exerccio para aprender a fundo o que est por trs desse processo.

Nesse tutorial faremos um exemplo de PCA na mo, sempre conferindo as contas com Python e depois faremos o mesmo exemplo usando a biblioteca Scikit-learn.

## Pequeno exemplo de PCA para fazer na mo

Como utilizaremos o Python para conferir nossas contas, importaremos as bibliotecas necessrias.

```python
# bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Utilizaremos as seguintes matrizes para esse trabalho:
<center>
$ X = \begin{bmatrix} 1 & 2 \\ 3 & 4 \\ 1 & 3 \\ 2 & 4 \\ 2 & 3 \\ 1 & 4 \end{bmatrix} $ &emsp; e &emsp; $ y = \begin{bmatrix} 0 \\ 1 \\ 0 \\ 1 \\ 1 \\ 1 \end{bmatrix} $
</center>

A matriz **X**  representa as features, enquanto **y**  representa o target. Declararemos essas matrizes no Python:

```python
# matrizes no Python
df = pd.DataFrame({
    'var1':[1,3,1,2,2,1],
    'var2':[2,4,3,4,3,4],
    'target': [0,1,0,1,1,1]
})
X = df.drop('target',1)
y = df['target']
```

Precisaremos das mdias de cada coluna de **X**. Para o clculo da mdia da coluna **j** basta fazer a seguinte conta:

$$\frac{\sum \limits _{i=1} ^{N} X_{ij}}{n}$$

com **n** sendo o nmero de linhas da matriz **X**. Podemos conferir as contas no Python usando a biblioteca numpy:

```python
# mdia por coluna de X
mean_vec = np.mean(X, axis=0) ## axis=0 --> coluna
print(mean_vec)
```

Nossa prpria tarefa  determinar a matriz de covarincia.

Para isso, vamos usar uma matriz **M** auxiliar.

Para determinar **M** subtraia a mdia da coluna **i** da coluna respectiva. 

Como no nosso exemplo todas as entradas de **M** esto multiplicadas por $\frac{1}{3}$ , para ficar mais elegante, colocamos $\frac{1}{3}$ em evidncia.

$$
M = \Large\frac{1}{3}\normalsize\begin{bmatrix} -2 & -4 \\ 4 & 2 \\ -2 & -1 \\ 1 & 2 \\ 1 & -1 \\ -2 & 2 \end{bmatrix}
$$

Para realizar a mesma conta no Python, use o cdigo a seguir:

```python
# subtraindo a mdia da respectiva coluna de X
M = X - mean_vec
```

A matriz de covarincia  de  X   dada por:
$$
C=\frac{1}{n1}M^{T}M
$$ 
onde  n   o nmero de linhas de  X.

Logo, $C=\Large\frac{1}{3}\normalsize\begin{bmatrix} 2 & 1 \\ 1 & 2 \end{bmatrix}$
 
No Python:

```python
# calculando matriz de covarincia
C = M.T.dot(M) / (X.shape[0]-1)
```

```python
# imprimindo C
C
```

Prximo passo  determinar auto-valores e auto-vetores da matriz de covarincia C. Os auto-valores $\lambda$ de C so raizes de
$pC(x)=det(CxI)$
 
onde I  a matriz identidade.

Cada auto-valor $\lambda$ tem auto-vetores associados. Apesar de auto-vetores nunca serem nulos, se inclurmos o vetor nulo no conjunto dos auto-vetores associados e chamarmos esse novo conjunto de $V_\lambda$, temos que $V_\lambda$  o conjunto soluo do seguinte sistema :$(C\lambda I)X=0$
 
Tomaremos geradores unitrios de $V\lambda 1$ e de $V\lambda 2$, para $\lambda 1=1$ e $\lambda 2= \frac{1}{3}$, a saber, tomaremos os seguintes auto-vetores:

$$
\left\{ \Bigg( \frac{ \sqrt{2}}{2}, \frac{ \sqrt{2}}{2}\Bigg),\Bigg( \frac{ \sqrt{2}}{2}, \frac{ \sqrt{2}}{2}\Bigg) \right\}
$$

Essa escolha no  nica e isso pode dar pequenas diferenas nos resultados finais.

No Python:

```python
# determinando auto-valores e auto-vetores
autovalores, autovetores = np.linalg.eig(C)

# imprimindo auto-valores de C
print("Auto-valores:")
print(autovalores)
print()

# imprimindo autovetores
print("Auto-vetores:")
print(autovetores)
```

Agora, ordenaremos os auto-valores do maior para o menor. Assim $\lambda 1=1$ e $\lambda 2 =\frac{1}{3}$.
 
No Python:

```python
# ordenando em ordem descrecente
pares_de_autos = [
    (
        np.abs(autovalores[i]),
        autovetores[:,i]
    ) for i in range(len(autovalores))
]
pares_de_autos.sort()
pares_de_autos.reverse()
```

Nosso prximo passo  calcular o quanto cada componente (auto-vetores) esto representando (de certa forma) a variabilidade dos nossos dados. Para isso usaremos duas medidas.

A varincia explicada de cada auto-valor $\lambda$  dada por:
$$
\large\frac{\lambda}{\sum _{j} \lambda_{j}}
$$
 
A varincia explicada acumulada de cada auto-valor $\lambda _{i}$   dada por:
$$
\frac{\sum _{j \leq i} \lambda _{j}} {\sum _{j} \lambda_{j}}
$$

No Python:

```python
# calculando a varincia explicada e a varincia explicada cumulativa
total = sum(autovalores)
var_exp = [
    (i / total)*100 for i in sorted(
        autovalores, reverse=True
    )
]
cum_var_exp = np.cumsum(var_exp)
```


Vejamos a forma sumarizada at o momento:


```python
# visualizando as informaes
x = [
    'PC %s' %i for i in range(
        1,len(autovalores)+1
    )
]
df_temp = pd.DataFrame(
    {'auto-valores': autovalores,
     'cum_var_exp':cum_var_exp,
     'var_exp':var_exp,
     'Componente':x}
)
print(df_temp)
print()
print("Auto-vetores")
for autovetor in [p[1] for p in pares_de_autos]:
    print(autovetor)
print()
```

Bom, isso  tudo!

Considerando os dados originais como pares ordenados em $\mathbb{R}^{2}$ , pois  $X$  possui apenas duas colunas, ao mudar da base cannica para a base composta pelos auto-vetores encontrados, teremos uma nova representao do conjunto $X$.

Vejamos graficamente:

```python
# visualizando graficamente os dados originais
sns.pairplot(
    df,
    vars = ['var1','var2'],
    hue='target',
    diag_kind="hist"
)
plt.show()
```

```python
# visualizando graficamente os dados atravs da mudana de base
# considerando base de auto-vetores
n_componentes = 2 # projetamos nas 2 (duas) primeiras componentes
autovetores = [p[1] for p in pares_de_autos]
A = autovetores[0:n_componentes] 
X = np.dot(X,np.array(A).T)
new_df = pd.DataFrame(X,columns=['pc1','pc2'])
new_df['target'] = df['target']
sns.pairplot(
    new_df, vars = ['pc1','pc2'], hue='target', diag_kind="hist"
)
plt.show()
```

* Os dados originais foram rotacionados por um ngulo de 45 uma vez que os auto-vetores so perpendiculares e obtidos dos eixos x e y por uma rotao de 45 no sentido anti-horrio.

* Nossos dados possuiam 2 colunas, mas poderiam possuir bem mais. Alm disso escolhermos 2 componentes para visualizao grfica, esse nmero tambm pode ser mais, desde que seja menor que o nmero de colunas.

* A ideia do PCA  diminuir a dimenso do conjunto de dados original escolhendo um nmero menor de componentes para projeo sobre estas.

* O processo de reduo de dimenso facilita processamento e armazenamento de dados.

## Fazendo uso da biblioteca

Agora, vejamos o mesmo exemplo utilizando a Scikit-learn.

```python
# biblioteca
from sklearn.decomposition import PCA

# dados
df = pd.DataFrame({
    'var1':[1,3,1,2,2,1],
    'var2':[2,4,3,4,3,4],
    'target': [0,1,0,1,1,1]
})
X = df.drop('target',1)
y = df['target']

# pca
pca = PCA(n_components=2)
pca.fit(X)
```

Isso  tudo, agora vejamos as informaes para fins de comparao.

```python
print("Auto-valores:")
print(pca.explained_variance_)
print()

print("Auto-vetores:")
print(pca.components_)
print()

print("Varincia explicada:")
print(pca.explained_variance_ratio_)
print()
```

Observe que, como a escolha dos auto-vetores no  nica, o algoritmo implementado na biblioteca Scikit-learn escolheu auto-vetores diferentes dos nossos. 

Mas, isso no  problema, pois s houve inverso de sinal. 

Logo, quando fizermos a visualizao grfica da mudana de coordenada, haver uma inverso da figura. 

Vejamos:


```python
# Tranformando X
X = pca.transform(X)

# Visualizando
new_df = pd.DataFrame(X, columns=['pc1','pc2'])
new_df['target'] = df['target']
sns.pairplot(
    new_df, vars = ['pc1','pc2'], hue='target', diag_kind="hist")
plt.show()
```

Perfeito!
Mas essa tcnica serve para reduzir dimenses. 
Como nosso exemplo  apenas para fins educacionais,  hora de aprender com exemplos reais como usar o poderoso PCA!

## PCA no conjunto de dados Iris
No prximo exemplo utilizaremos PCA para reduzir o conjunto de dados Iris de 4 para 2 features. 

O conjunto pode ser baixado no repositrio UCI. 

O link encontra-se nas referncias.

```python
# dados
df = pd.read_csv('iris.data',header=None)
df.columns = [
    'sepal length in cm',
    'sepal width in cm',
    'petal length in cm',
    'petal width in cm',
    'target'
]
X = df.drop('target',1)
y = df['target']
```


Faremos novamente passo a passo. O cdigo  anlogo ao cdigo usado no exemplo anterior.


```python
mean_vec = np.mean(X, axis=0)
M = X - mean_vec
C = M.T.dot(M) / (X.shape[0]-1)
autovalores, autovetores = np.linalg.eig(C)
pares_de_autos = [
    (
        np.abs(autovalores[i]),
        autovetores[:,i]
    ) for i in range(len(autovalores))
]
pares_de_autos.sort()
pares_de_autos.reverse()
total = sum(autovalores)
var_exp = [
    (i / total)*100 for i in sorted(autovalores, reverse=True)
]
cum_var_exp = np.cumsum(var_exp)
```

```python
x = ['PC %s' %i for i in range(1,len(autovalores)+1)]
df_temp = pd.DataFrame(
    {'auto-valores': autovalores,
     'cum_var_exp':cum_var_exp,
     'var_exp':var_exp,
     'Componente':x}
)
print(df_temp)
print()
print("Auto-vetores")
for autovetor in [p[1] for p in pares_de_autos]:
    print(autovetor)
print()
```


Com duas componentes temos uma varincia explicada cumulativa de 97%. Utilizaremos ento, apenas duas componentes.


```python
n_componentes = 2 # projetamos nas 2 (duas) primeiras componentes
autovetores = [p[1] for p in pares_de_autos]
A = autovetores[0:n_componentes] 
X = np.dot(X,np.array(A).T)
new_df = pd.DataFrame(X, columns=['pc1','pc2'])
new_df['target'] = df['target']
sns.pairplot(
    new_df, vars = ['pc1','pc2'],
    hue='target', diag_kind="hist")
plt.show()
```

## PCA para reduo de dimenso em imagens

Outra aplicao  a reduo de dimenso em imagens.

Para esse exemplo, escolhemos a imagem Pssaro Mineiro Barulhento retirada do site Pixabay.

A imagem foi transformada em escala de cinza utilizando o software GIMP.

<div>

</div>

Primeiro, carregaremos a imagem.

```python
# carregando a imagem e visualizando dimenses
X = plt.imread('./images/bird.jpg')
X.shape
```

Observe que a matriz tem 1920 linhas e 1899 colunas. Aplicaremos, em seguida, a tcnica PCA.

```python
# aplicando PCA
pca = PCA(0.99) # varincia explicada de 0.99
lower_dimension_data = pca.fit_transform(X)
lower_dimension_data.shape
```

O nmero 0.99 passado como parmetro para o mtodos PCA diz que queremos um nmero de componentes que nos garanta 99% de varincia explicada cumulativa. 

Observe que para isso, utilizou-se 145 componentes. Isso  uma reduo imensa, uma vez que a original possuia 1899 colunas. 

Queremos ver, caso necessrio recuperar a imagem, como essas imagens ficam com a dimenso reduzida. 

Para isso sero necessrios duas funes auxiliares.

```python
def pca_with_var_exp(X, var_exp=0.99):
    pca = PCA(var_exp) # varincia explicada de 0.99
    lower_dimension_data = pca.fit_transform(X)
    print(lower_dimension_data.shape)
    approximation = pca.inverse_transform(lower_dimension_data)
    return approximation

def plot_subplot(X, i):
    plt.subplot(3,2,i)
    plt.imshow(X, cmap="gray")
    plt.xticks([])
    plt.yticks([])
```

A primeira funo reduz a imagem X mantendo var_exp de varincia explicada cumulativa. Compararemos 3 valores: 0.99, 0.95 e 0.90.

```python
# calculando algumas aproximaes
img_1 = pca_with_var_exp(X, var_exp=0.99)
img_2 = pca_with_var_exp(X, var_exp=0.95)
img_3 = pca_with_var_exp(X, var_exp=0.90)
```

A segunda funo mostra o grfico. Utilizaremos ela para comparar as imagens obtidas com a original.

```python
# visualizando aproximaes
plt.figure(figsize=(6,8))

plot_subplot(X, 1)
plt.title("Original")
plot_subplot(img_1, 2)
plt.title("Var. Explicada de 99%")
plot_subplot(X, 3)
plt.title("Original")
plot_subplot(img_2, 4)
plt.title("Var. Explicada de 95%")
plot_subplot(X, 5)
plt.title("Original")
plot_subplot(img_3, 6)
plt.title("Var. Explicada de 90%")

plt.tight_layout()
plt.show()
```
